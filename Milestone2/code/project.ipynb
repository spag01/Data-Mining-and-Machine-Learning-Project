```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, OrdinalEncoder
import csv

```

```python
test_cases_data = pd.read_csv("cases_2021_test_processed_unlabelled_2.csv")
test_cases_data
```

```python
train_cases_data = pd.read_csv("cases_2021_train_processed_2.csv")
train_cases_data
```

```python
# 1.1 Feature Selection

# Dropping Columms longitude, latitude, and date_confirmation
# Filling missing values with "No province" in the province column

# For train dataset
train_cases_data.province = train_cases_data.province.fillna('No Province')
train_cases_data = train_cases_data.drop(['latitude','longitude','date_confirmation'],axis=1) #Later inpace =True

# For test dataset
test_cases_data.province = test_cases_data.province.fillna('No Province')
test_cases_data = test_cases_data.drop(['latitude','longitude','date_confirmation'],axis=1) #Later inpace =True
```

```python
# 1.2 Mapping the Features
data_enc = LabelEncoder() 
data_enc1 = MinMaxScaler()
train_cases_data.columns

#For train dataset

# Using LabelEncoder on Categorical Features
train_cases_data['sex']=data_enc.fit_transform(train_cases_data['sex'])
train_cases_data['province']=data_enc.fit_transform(train_cases_data['province'])
train_cases_data['country']=data_enc.fit_transform(train_cases_data['country'])
train_cases_data['chronic_disease_binary']=data_enc.fit_transform(train_cases_data['chronic_disease_binary'])
train_cases_data['outcome_group']=data_enc.fit_transform(train_cases_data['outcome_group'])

# Using MinMaxScaler for Numerical Features
data_enc1.fit(train_cases_data[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'Case_Fatality_Ratio']])
train_cases_data[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'Case_Fatality_Ratio']] = data_enc1.transform(train_cases_data[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'Case_Fatality_Ratio']])


# For test dataset 

# Using LabelEncoder on Categorical Features
test_cases_data['sex']=data_enc.fit_transform(test_cases_data['sex'])
test_cases_data['province']=data_enc.fit_transform(test_cases_data['province'])
test_cases_data['country']=data_enc.fit_transform(test_cases_data['country'])
test_cases_data['chronic_disease_binary']=data_enc.fit_transform(test_cases_data['chronic_disease_binary'])

# Using MinMaxScaler for Numerical Features
data_enc1.fit(test_cases_data[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'Case_Fatality_Ratio']])
test_cases_data[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'Case_Fatality_Ratio']] = data_enc1.transform(test_cases_data[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'Case_Fatality_Ratio']])

# if using OrdinalEncoder
#train_cases_data[['sex', 'province', 'country','chronic_disease_binary','outcome_group']] = data_enc.transform(train_cases_data[['sex', 'province', 'country','chronic_disease_binary', 'outcome_group']])
```

```python
y = train_cases_data.drop(['age', 'sex', 'province', 'country', 'chronic_disease_binary',
      'Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate',
       'Case_Fatality_Ratio'], axis=1)

X = train_cases_data[['age', 'sex', 'province', 'country', 'chronic_disease_binary',
     'Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate',
      'Case_Fatality_Ratio']]
```

```python
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(
        X, y, train_size=0.8,test_size=0.2, random_state=100, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.metrics import accuracy_score
rf = RFC()
rf.fit(X_train,y_train.values.ravel())
y_pred = rf.predict(X_val)
print('Test accuracy score:',rf.score(X_val,y_val))
print('Train accuracy score:',rf.score(X_train,y_train))
print('Accuracy score:',accuracy_score(y_val,y_pred))
```

```python
from sklearn.metrics import classification_report
print("classification report before balancing the data with RandomForestClassifier:" , "\n")
print(classification_report(y_val,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train.values.ravel())
y_pred = knn.predict(X_val)
print('Test score:',knn.score(X_val,y_val))
print('Train score:',knn.score(X_train,y_train))
print('Accuracy score:',accuracy_score(y_val,y_pred))
```

```python
from sklearn.metrics import classification_report
print("classification report before balancing the data with kNeighborsClassifier:" , "\n")
print(classification_report(y_val,y_pred))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(max_depth = 4, gamma =0.1)
xgb.fit(X_train, y_train.values.ravel())
y_pred = xgb.predict(X_val)
print('Test score:',xgb.score(X_val, y_val))
print('Train score:',xgb.score(X_train, y_train))
print('Accuracy score:',accuracy_score(y_val,y_pred))
```

```python
print("classification report before balancing the data with XGBClassifier :" , "\n")
print(classification_report(y_val,y_pred))
```

```python

#1.4 Hyperparamter Tuning Using RandomizedSearchCV and GridSearchCV

from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import f1_score, make_scorer 

n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt','log2']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10,1000,10)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10,14]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4,6,8]
# Create the random grid

rf_param = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
              'criterion':['entropy','gini']
            }

rf_scoring = {'accuracy': make_scorer(accuracy_score),
              'macro_f1':make_scorer(f1_score, average='macro'), 
              'f1_deceased':make_scorer(f1_score,labels=[0],average='macro'),
              'f1_hospitalized': make_scorer(f1_score,labels=[1],average='macro'),
              'f1_nonhospitalized': make_scorer(f1_score,labels=[2],average='macro'),
              }
              
print(rf_param)
```

```python
rf_randomcv = RandomizedSearchCV(estimator=rf , param_distributions=rf_param, n_iter=20, cv=5, verbose=2,
                               random_state=100, n_jobs=-1, scoring = rf_scoring, refit='accuracy')
# fit to randomized rf
rf_randomcv.fit(X_bal,y_bal.values.ravel())
```

```python
# Printing the best hyperparameter combination for the model
print(rf_randomcv.best_params_ ,"\n")

# to make new classifier with the best estimator
best_rf_param = rf_randomcv.best_estimator_
#print(best_rf_param)
print(rf_randomcv.best_score_)
```

```python
#1.5 Overfitting => Choosing one hyperparameter and changing it's value to see change in test and train accuarcies
```

```python
from matplotlib import pyplot
train_scores, test_scores = list(), list()
values = [i for i in range(1, 21)]
for i in values:
    model = RFC(max_depth=i)
    model.fit(X_train, y_train.values.ravel())
    train_yhat = model.predict(X_train)
    train_acc = accuracy_score(y_train, train_yhat)
    train_scores.append(train_acc)
    test_yhat = model.predict(X_val)
    test_acc = accuracy_score(y_val, test_yhat)
    test_scores.append(test_acc)
    print('>%d, train: %.3f, test: %.3f' % (i, train_acc, test_acc))

pyplot.plot(values, train_scores, '-o', label='Train')
pyplot.plot(values, test_scores, '-o', label='Test')
pyplot.legend()
pyplot.show()
```

```python
train_scores1, test_scores1 = list(), list()
values = [i for i in range(1, 21)]
for i in values:
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train, y_train.values.ravel())
    train_yhat = model.predict(X_train)
    train_acc = accuracy_score(y_train, train_yhat)
    train_scores1.append(train_acc)
    test_yhat = model.predict(X_val)
    test_acc = accuracy_score(y_val, test_yhat)
    test_scores1.append(test_acc)
    print('>%d, train: %.3f, test: %.3f' % (i, train_acc, test_acc))

pyplot.plot(values, train_scores1, '-o', label='Train')
pyplot.plot(values, test_scores1, '-o', label='Test')
pyplot.legend()
pyplot.show()
```

```python
train_scores2, test_scores2 = list(), list()
values = [i for i in range(1, 21)]
for i in values:
    model = XGBClassifier(max_depth=i)
    model.fit(X_train, y_train.values.ravel())
    train_yhat = model.predict(X_train)
    train_acc = accuracy_score(y_train, train_yhat)
    train_scores2.append(train_acc)
    test_yhat = model.predict(X_val)
    test_acc = accuracy_score(y_val, test_yhat)
    test_scores2.append(test_acc)
    print('>%d, train: %.3f, test: %.3f' % (i, train_acc, test_acc))

pyplot.plot(values, train_scores2, '-o', label='Train')
pyplot.plot(values, test_scores2, '-o', label='Test')
pyplot.legend()
pyplot.show()
```

```python
# Getting the csv file with our prediction values predicted by best performing classification model 

def create_submission_file(y_preds, file_name): 
  with open(file_name, 'w') as csvfile:
    wr = csv.writer(csvfile, quoting=csv.QUOTE_ALL)
    wr.writerow(["Id", "Prediction"])
    for i, pred in enumerate(y_preds): 
      wr.writerow([str(i), str(pred)])
      
create_submission_file(y_preds, "submission_RandomForestClassifier.csv")
```
